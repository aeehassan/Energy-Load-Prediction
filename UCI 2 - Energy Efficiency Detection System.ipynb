# Data preprocessing

## Standardization

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
model_inputs = eed_df.drop(['Y1','Y2'], axis=1).columns
eed_df[model_inputs] = ss.fit_transform(eed_df[model_inputs])
eed_df
```

```python
eed_df.describe().round(3)
```

## Data splitting

```python
from sklearn.model_selection import train_test_split
```

```python
y = eed_df[["Y1","Y2"]]
X = eed_df.drop(["Y1","Y2"], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

## Model building

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model_two = LinearRegression()
lr_model_two.fit(X_train, y_train)
```

## Evaluation

```python
y_pred = lr_model_two.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error, mean_squared_error
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
rmse = root_mean_squared_error(y_test, y_pred)
rmse
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
# Rule of thumb check 
rmse > 0.2 * y.max()
```
